leetcode gym schedule by patterned number of problems a day

```python
target = int(input())
multiplier = int(input("1 is hardest to 10 is the easiest: "))

total = 0
count = 0
current = 1
times_used = 0

for _ in range(target + 1):  # Upper bound - worst case every addition is 1
    if total >= target:
        break
    
    total += current
    count += 1
    times_used += 1
    
    if times_used == (multiplier * current):
        current += 1
        times_used = 0

print(f"Total: {total}, Last number: {current - 1 if times_used == 0 else current}, Count: {count}, Iteration: {times_used}")
```
